# H.2 Ethical Considerations

While predictive models and segmentation can guide useful interventions, several ethical issues must be considered:

1. **Privacy of Student Data**  
   - Sensitive academic and behavioral data must be handled with strict **data governance and anonymization**.  
   - Mitigation: Limit access, apply **secure storage** and **data minimization** principles.

2. **Fairness & Bias**  
   - Models may overestimate risk for students with many absences (e.g., due to health or family circumstances) or under-support those from disadvantaged backgrounds.  
   - Mitigation: Conduct **fairness audits** across gender, socioeconomic status, and other attributes.

3. **Avoiding Stigmatization**  
   - Labeling students as ‚Äúhigh-risk‚Äù can harm self-esteem.  
   - Mitigation: Frame outputs as **support needs**, not failure predictions.

4. **Transparency**  
   - Teachers, parents, and students should understand how predictions are made.  
   - Mitigation: Prefer **interpretable models (logistic coefficients, feature importances)** and **clear reporting**.

5. **Responsible Use of Prior Grades (G1/G2)**  
   - While predictive, relying heavily on prior grades risks reinforcing disadvantage.  
   - Mitigation: Use G1/G2 **in combination with behavioral signals** to ensure early interventions, not just reactive measures.
